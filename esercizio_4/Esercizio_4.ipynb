```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

plt.rcParams.update({'font.size': 13})
```

```python
k_B = 1.380649 * 10**(-23) # J/K
epsilon = 120 * k_B # J
m = 39.948*1.66054*10**(-27) # g
sigma = 0.34 # nm
t = sigma*np.sqrt(m/epsilon)
```

### Argon in Fase Solida

```python
# Instantaneous Observables
plt.figure(figsize=(12,7))
datafile = "./data/4/solid/solid.temp.dat"
val = np.loadtxt(datafile)
N = 100000
x = np.arange(N)*t
plt.plot(x, val, label='Temperature')

datafile = "./data/4/solid/solid.etot.dat"
val = np.loadtxt(datafile)
plt.plot(x, val, label='Total Energy')

datafile = "./data/4/solid/solid.epot.dat"
val = np.loadtxt(datafile)
plt.plot(x, val, label='Potential Energy')

datafile = "./data/4/solid/solid.ekin.dat"
val = np.loadtxt(datafile)
plt.plot(x, val, label='Kinetic Energy')

plt.title("Solid Phase Instantaneous Measurements")
plt.xlabel("time (s)")
plt.ylabel("Energy (LJU)")
plt.legend()
plt.grid(True)
```

```python
# Blocked Averages
fig, ax = plt.subplots(2, 2, figsize=(14, 8))
fig.suptitle("Solid Phase (LJU)")

N = 100  # Blocks
x = np.arange(N)

datafile = "./data/4/solid/solid.ave_etot.dat"
avg, err = np.loadtxt(datafile, usecols=(1, 2), delimiter=' ', unpack='true')
ax[0, 0].errorbar(x, avg, yerr=err, label="etot")
ax[0, 0].title.set_text("Total Energy (LJU)")
ax[0, 0].grid(True)

datafile = "./data/4/solid/solid.ave_epot.dat"
avg, err = np.loadtxt(datafile, usecols=(1, 2), delimiter=' ', unpack='true')
ax[0, 1].errorbar(x, avg, yerr=err,  label="epot")
ax[0, 1].title.set_text("Potential Energy (LJU)")
ax[0, 1].grid(True)

datafile = "./data/4/solid/solid.ave_ekin.dat"
avg, err = np.loadtxt(datafile, usecols=(1, 2), delimiter=' ', unpack='true')
ax[1, 0].errorbar(x, avg, yerr=err, label="ekin")
ax[1, 0].title.set_text("Kinetic Energy (LJU)")
ax[1, 0].grid(True)

datafile = "./data/4/solid/solid.ave_temp.dat"
avg, err = np.loadtxt(datafile, usecols=(1, 2), delimiter=' ', unpack='true')
ax[1, 1].errorbar(x, avg, yerr=err, label="temp (K)")
ax[1, 1].title.set_text("Temperature (LJU)")
ax[1, 1].grid(True)

plt.tight_layout()
plt.show()
```

```python
# SI Units?
fig, ax = plt.subplots(2, 2, figsize=(14, 8))
fig.suptitle("Solid Phase (SI)")

N = 100  # Blocks
x = np.arange(N)

scale = 120

datafile = "./data/4/solid/solid.ave_etot.dat"
avg, err = np.loadtxt(datafile, usecols=(1, 2), delimiter=' ', unpack='true')
ax[0, 0].errorbar(x, avg*scale, yerr=err*scale, label="etot")
ax[0, 0].title.set_text("Total Energy (J * k_B)")
ax[0, 0].grid(True)

datafile = "./data/4/solid/solid.ave_epot.dat"
avg, err = np.loadtxt(datafile, usecols=(1, 2), delimiter=' ', unpack='true')
ax[0, 1].errorbar(x, avg*scale, yerr=err*scale,  label="epot")
ax[0, 1].title.set_text("Potential Energy (J * k_B)")
ax[0, 1].grid(True)

datafile = "./data/4/solid/solid.ave_ekin.dat"
avg, err = np.loadtxt(datafile, usecols=(1, 2), delimiter=' ', unpack='true')
ax[1, 0].errorbar(x, avg*scale, yerr=err*scale, label="ekin")
ax[1, 0].title.set_text("Kinetic Energy (J * k_B)")
ax[1, 0].grid(True)

datafile = "./data/4/solid/solid.ave_temp.dat"
avg, err = np.loadtxt(datafile, usecols=(1, 2), delimiter=' ', unpack='true')
ax[1, 1].errorbar(x, avg*scale, yerr=err*scale, label="temp (K)")
ax[1, 1].title.set_text("Temperature (K)")
ax[1, 1].grid(True)

plt.tight_layout()
plt.show()
```
