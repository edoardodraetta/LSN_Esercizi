### 2.1.1 Basic Monte Carlo Integration

```python

datafile = "./rng/data/stats_2.1.1.dat"
I, err = np.loadtxt(datafile, usecols=(0,2), delimiter=' ', unpack='true')

M = 1000000    # Throws
N = 100        # Blocks
L = int(M/N)   # Throws per block

x = np.arange(N)
x *= L  # coordinates of each error bar
plt.figure(figsize=(10, 5))
plt.grid(True)
plt.errorbar(x, I - 1, yerr=err)

plt.xlabel('# throws')
plt.ylabel(r'$I- 1 $ ')
plt.title(r'Monte Carlo Integration of $\int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx$ ')
plt.savefig('./figures/2.1.1.png')
plt.show()

plt.plot(x, err)
```
