### 2.1.1 Basic Monte Carlo Integration

```python
datafile = "./rng/data/stats_2.1.1.dat"
I, err = np.loadtxt(datafile, usecols=(0,2), delimiter=' ', unpack='true')

M = 1000000    # Throws
N = 100        # Blocks
L = int(M/N)   # Throws per block

x = np.arange(N)
x *= L  # coordinates of each error bar
plt.figure(figsize=(10, 5))
plt.grid(True)
plt.errorbar(x, I - 1, yerr=err)

plt.xlabel('# throws')
plt.ylabel(r'$I- 1 $ ')
plt.title(r'Monte Carlo Integration of $\int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx$ ')
plt.savefig('./figures/2.1.1.png')
plt.show()

plt.plot(x, err)
```

### 2.2.1 Random Walks on a Lattice

```python
from scipy.optimize import curve_fit

def f(x,k):  # defining the fitting function
    return k * np.sqrt(x)

datafile = "./rng/data/stats_2.2.1.dat"
dist, err = np.loadtxt(datafile, usecols=(0,2), delimiter=' ', unpack='true')

n_walks = 100       
n_steps = 10000 

x = np.arange(n_walks)
plt.figure(figsize=(10, 5))
plt.grid(True)
plt.errorbar(x, dist, yerr=err)

p_opt, p_cov = curve_fit(f, x, dist, bounds=(-5,5))
y_fit = f(x,p_opt)
plt.plot(x,y_fit) # plotting fitted function
print(p_opt)

plt.xlabel('# steps')
plt.ylabel('Distance From Origin')
plt.title('Discrete Random Walks')
plt.savefig('./figures/2.2.1.png')
plt.show()
```
