```python
import matplotlib
import matplotlib.pyplot as plt
# from mpl_toolkits.mplot3d import Axes3D
from matplotlib import gridspec
import numpy as np
import math
from scipy.optimize import curve_fit

plt.rcParams.update({'font.size': 12})
```

```python
datafile = "./data/stats_3.1.1_call.dat"
call, call_err = np.loadtxt(datafile, usecols=(0, 2), delimiter=' ', unpack='true')

datafile = "./data/stats_3.1.1_put.dat"
put, put_err = np.loadtxt(datafile, usecols=(0, 2), delimiter=' ', unpack='true')

N = 100
x = np.arange(N)
plt.figure(figsize=(10, 5))
plt.grid(True)

plt.errorbar(x, call, yerr=call_err)

plt.xlabel('# Blocks')
plt.ylabel('Price' )
plt.title('Call Option Price (Direct)')
plt.savefig('./figures/3.1.1_call.png')
plt.show()

plt.figure(figsize=(10, 5))
plt.grid(True)

plt.errorbar(x, put, yerr=put_err)

plt.xlabel('# Blocks')
plt.ylabel('Price' )
plt.title('Put Option Price (Direct)')
plt.savefig('./figures/3.1.1_put.png')
plt.show()
```
