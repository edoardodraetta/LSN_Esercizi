### Exercise 03.1 <span style="color:red"> Plain vanilla option pricing </span>

Qui sono riportati i risultati di simulazioni per il calcolo dei prezzi di opzioni finanziarie usando la soluzione analita per il moto browniano geometrico $GBM(r,\sigma^2)$.

Le simulazioni sono state eseguite in due modi diversi: usando la soluzione analitica per il prezzo $S(t)$ campionata direttamente al tempo $T$, e usando $S(t)$ per portarci con tanti piccoli passi dal tempo attuale $t=0$ al tempo $T$. Entrambi risultati convergono abbastanza bene al risultato ottenuto con l'equazione di Black-Scholes. 

Quale risultato è più affidabile? Quali parametri sono stati utilizzati?

```python
datafile = "./data/stats_3.1.1_call.dat"
call, call_err = np.loadtxt(datafile, usecols=(0, 2), delimiter=' ', unpack='true')

datafile = "./data/stats_3.1.1_put.dat"
put, put_err = np.loadtxt(datafile, usecols=(0, 2), delimiter=' ', unpack='true')

N = 100

fig, ax = plt.subplots(1, 2, figsize=(18, 6), sharex=True)
fig.suptitle("Options Pricing: Direct Sampling of GBM", fontsize=20, y=1.05)

x = np.arange(N)

ax[0].errorbar(x, call, yerr=call_err)
ax[0].grid(True)
ax[0].title.set_text("Call Option")
ax[0].set_ylabel("Price")
ax[0].set_xlabel("# Blocks")

ax[1].errorbar(x, put, yerr=put_err)
ax[1].title.set_text("Put Option")
ax[1].set_xlabel("# Blocks")
ax[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
datafile = "./data/stats_3.1.2_call.dat"
call, call_err = np.loadtxt(datafile, usecols=(0, 2), delimiter=' ', unpack='true')

datafile = "./data/stats_3.1.2_put.dat"
put, put_err = np.loadtxt(datafile, usecols=(0, 2), delimiter=' ', unpack='true')

N = 100

fig, ax = plt.subplots(1, 2, figsize=(18, 6), sharex=True)
fig.suptitle("Options Pricing: Path Sampling of GBM", fontsize=20, y=1.05)

x = np.arange(N)

ax[0].errorbar(x, call, yerr=call_err)
ax[0].grid(True)
ax[0].title.set_text("Call Option")
ax[0].set_ylabel("Price")
ax[0].set_xlabel("# Blocks")

ax[1].errorbar(x, put, yerr=put_err)
ax[1].title.set_text("Put Option")
ax[1].set_xlabel("# Blocks")
ax[1].grid(True)

plt.tight_layout()
plt.show()
```
