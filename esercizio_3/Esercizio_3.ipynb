```python
import matplotlib
import matplotlib.pyplot as plt
# from mpl_toolkits.mplot3d import Axes3D
from matplotlib import gridspec
import numpy as np
import math
from scipy.optimize import curve_fit

plt.rcParams.update({'font.size': 14})
```

```python
datafile = "./data/stats_3.1.1.dat"
avg, err = np.loadtxt(datafile, usecols=(0, 2), delimiter=' ', unpack='true')

M = 1000000    # Throws
N = 100        # Blocks
L = int(M/N)   # Throws per block

x = np.arange(N)
x *= L  # coordinates of each error bar
plt.figure(figsize=(10, 5))
plt.grid(True)
plt.errorbar(x, avg, yerr=err)

plt.xlabel('# throws')
plt.ylabel('Price' )
plt.title('Put Option price (Direct)')
plt.savefig('./figures/3.1.1.png')
plt.show()
```
